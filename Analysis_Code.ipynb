# **Data Pre-Processing**

All the data preprocessing approaches are finalized during EDA. Now we are going to implement these.

```python
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import MinMaxScaler
import numpy as np

#Applying the simple imputers
imputer = SimpleImputer(strategy='median', missing_values=np.nan)
imputer = imputer.fit(df_eda[['Age']])
df_eda[['Age']]= imputer.transform(df_eda[['Age']])
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(df_eda[['HasDrivingLicense']])
df_eda[['HasDrivingLicense']]= imputer.transform(df_eda[['HasDrivingLicense']])

df_clean = df_eda.copy()
df_clean = df_clean.dropna(axis = 0, subset=['RegionID'])
df_clean = df_clean.drop('id',axis = 1)

print(df_clean.isna().sum())
print(df_clean.describe())
df_clean['AnnualPremium'] = df_clean['AnnualPremium'].str.replace(',', '').str.replace('£', '')
```

```python
#Applying the KNN imputers
# Going to use following method to apply KNNImputer.
# First going to label the categorical features, then
# apply knn imputer. At the end reverse the labeling to get the dataset.
imputer_knn = KNNImputer(n_neighbors=1, weights='uniform', metric='nan_euclidean')
mm = MinMaxScaler()
cols_to_impute = ['Gender','Switch','VehicleAge','PastAccident']
mappin = dict()
for variable in cols_to_impute:
  mappings = {k: i for i, k in enumerate(df_clean[variable].dropna().unique(), 0)}
  mappin[variable] = mappings

for variable in cols_to_impute:
  df_clean[variable] = df_clean[variable].map(mappin[variable])

#Minmaxscaler and KNN imputation
sca = mm.fit_transform(df_clean)
knn_imputer = KNNImputer()
knn = knn_imputer.fit_transform(sca)
df_clean.iloc[:,:] = mm.inverse_transform(knn)
for i in df_clean.columns :
  df_clean[i] = round(df_clean[i]).astype('int')

#Inverse transform
for i in cols_to_impute:
  inv_map = {v: k for k, v in mappin[i].items()}
  df_clean[i] = df_clean[i].map(inv_map)

print(df_clean.isna().sum())
```

# Preprocessing Pipeline and Column Transformer

```python
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None, 'max_colwidth', None, 'display.expand_frame_repr', False)
print(df_ready)

# split the data into input and output
df_ready_x = df_ready[['Gender', 'Age', 'HasDrivingLicense', 'RegionID',
       'VehicleAge',  'AnnualPremium', 'SalesChannelID',
       'DaysSinceCreated','Switch','PastAccident']]
df_ready_y = df_ready[['Result']]
```

```python
# As per the desciption provided with the assignment
numerical_features=['Age','AnnualPremium','DaysSinceCreated']
categorical_features=['Gender','HasDrivingLicense','RegionID','Switch','VehicleAge','PastAccident','SalesChannelID']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer

# transformer for categorical features
categorical_transformer = Pipeline(
    [
        ('onehot', OneHotEncoder(handle_unknown = 'ignore'))
    ]
)

# transformer for numerical features
numeric_transformer = Pipeline(
    [
        ('scaler', StandardScaler())
    ]
)

# combine them in a single ColumnTransformer
preprocessor = ColumnTransformer(
    [
        ('categoricals', categorical_transformer, categorical_features),
        ('numericals', numeric_transformer, numerical_features)
    ],
    remainder = 'drop'
)
```

```python
x_processed = preprocessor.fit_transform(df_ready_x)
X_train, X_test, y_train, y_test = train_test_split(df_ready_x, df_ready_y, test_size=0.3, random_state=25)
```

**SMOTE for Balancing** - During EDA it was concluded that target class is unbalanced(data unbalanced). Now we are going to apply smote to balance the data. This will be oversampling the lower target class data to balance the  target classes.

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

#Using Smote to oversample
X_train_S, X_test_S, y_train_S, y_test_S = train_test_split(x_processed, df_ready_y, test_size = 0.3, random_state= 25)
oversample = SMOTE()
X_smote, y_smote = oversample.fit_resample(X_train_S, y_train_S)
```

# **Model Implementation**

Now, we are going to implement 3 model as per the the coursework requirement. Since we are dealing with classification problem here. The 3 model used will be classifiers. We are going to use **XGBoost**, **Random Forest** and **Decision trees**. We are going to be running the analysis on both unbalanced and balanced data.

Since our data is imbalanced, we are going to focus on F1 Score in hyper parameter tuning of models. F1 score and PR curve is better metric for evaluation and optimization in this case.

Model Evaluator Methods

```python
from sklearn.metrics import accuracy_score, f1_score, precision_recall_curve, auc, mean_absolute_error
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, ConfusionMatrixDisplay
from matplotlib import pyplot

def evaluate(model, test_features, test_labels, model_name):
  predictions = model.predict(test_features)
  accuracy = accuracy_score(test_labels, predictions)
  y_score = model.predict_proba(test_features)[:,1]
  Roc_Auc_Score = roc_auc_score(test_labels, y_score)
  precision, recall, _ = precision_recall_curve(test_labels, y_score)
  PR_Auc_Score = auc(recall, precision)
  errors = mean_absolute_error(predictions, test_labels.T.values[0])
  print(model_name + ' Model Performance')
  print("ROC-AUC Score: ",Roc_Auc_Score)
  print("PR-AUC Score: ",PR_Auc_Score)
  print("F1 Score: ",f1_score(test_labels, predictions,average = None))
  print('Average Error: {:0.4f} degrees.'.format(errors))
  print('Accuracy = {:0.2f}%.'.format(accuracy*100))
  print(classification_report(test_labels, predictions))
  return accuracy, f1_score(test_labels, predictions,average = None), PR_Auc_Score, y_score, predictions

def plot_curve_roc(test_labels, y_score, model_name):
  # plotting ROC curve
  m_fpr, m_tpr, _ = roc_curve(test_labels, y_score)
  pyplot.figure(figsize=[6,6])
  pyplot.title("ROC Curve")
  pyplot.plot(m_fpr, m_tpr, label=model_name)
  # axis labels
  pyplot.xlabel('False Positive Rate')
  pyplot.ylabel('True Positive Rate')
  # show the legend
  pyplot.legend()
  # show the plot
  pyplot.show()
  print ('Area under curve (AUC): ', auc(m_fpr,m_tpr))


def plot_curve_pr(test_labels, y_score, model_name):
  # plotting precision-recall curve
  lr_precision, lr_recall, _ = precision_recall_curve(test_labels, y_score)
  pyplot.figure(figsize=[6,6])
  pyplot.title("Precision Recall Curve")
  # plot the precision-recall curves
  pyplot.plot(lr_recall, lr_precision, label=model_name)
  # axis labels
  pyplot.xlabel('Recall')
  pyplot.ylabel('Precision')
  # show the legend
  pyplot.legend()
  # show the plot
  pyplot.show()

def plot_compare_curve_roc(test_labels, y_score, model_name):
  # plotting ROC curve
  m_fprN, m_tprN, _ = roc_curve(test_labels[0], y_score[0])
  m_fprS, m_tprS, _ = roc_curve(test_labels[1], y_score[1])
  pyplot.figure(figsize=[6,6])
  pyplot.plot(m_fprN, m_tprN, label=model_name+" without SMOTE")
  pyplot.plot(m_fprS, m_tprS, label=model_name+" with SMOTE")
  # axis labels
  pyplot.xlabel('False Positive Rate')
  pyplot.ylabel('True Positive Rate')
  # show the legend
  pyplot.legend()
  # show the plot
  pyplot.show()
  print ('Area under curve (AUC) without SMOTE: ', auc(m_fprN,m_tprN))
  print ('Area under curve (AUC) with SMOTE: ', auc(m_fprS,m_tprS))


def plot_compare_curve_pr(test_labels, y_score, model_name):
  # plotting precision-recall curve
  lr_precisionN, lr_recallN, _ = precision_recall_curve(test_labels[0], y_score[0])
  lr_precisionS, lr_recallS, _ = precision_recall_curve(test_labels[1], y_score[1])
  pyplot.figure(figsize=[6,6])
  # plot the precision-recall curves
  pyplot.plot(lr_recallN, lr_precisionN, label=model_name+" without SMOTE")
  pyplot.plot(lr_recallS, lr_precisionS, label=model_name+" with SMOTE")
  # axis labels
  pyplot.xlabel('Recall')
  pyplot.ylabel('Precision')
  # show the legend
  pyplot.legend()
  # show the plot
  pyplot.show()

def show_confusion_matrix(actual, predicted):
  cm = confusion_matrix(actual, predicted)
  cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [False, True])
  cm_display.plot()
  plt.show()
```
